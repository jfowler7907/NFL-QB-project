```python
#Import dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
#Create path to draft CSV and convert to DataFrame by reading in the CSV
draft_path = 'resources/nfl_draft.csv'
draft_df = pd.read_csv(draft_path, encoding ='utf8')
draft_df
```

```python
#Drop all non quarterbacks fron the df and set this as a new df
#Filter the draft data to only inclde players drafted after 2004
#Select the player name as the index
draft_df_qb = draft_df.loc[draft_df['Pos']=='QB']
draft_df_qb = pd.DataFrame(draft_df_qb)
draft_df_qb.set_index('Player', inplace = True)
draft_df_qb= draft_df_qb[['Year','Rnd','Pick','Player_Id','Pos']]
draft_df_qb = draft_df_qb.loc[draft_df_qb['Year']>=2004]
draft_df_qb = draft_df_qb.drop('Player_Id', axis=1)
draft_df_qb
```

```python
#Create path to stats CSV and convert to DataFrame by reading in the CSV
#Filter out stats before 2004 to line up with draft data
#Reflect df
stats_path = 'resources/QBStats_all.csv'
stats_df = pd.read_csv(stats_path, encoding ='utf8')
stats_df = stats_df.loc[stats_df['year']>=2004]
stats_df.head()
```

```python
#Cleaned up the naming of the players in the Stats df to line up with the draft df
qb_stats_df = stats_df
qb_stats_df
names = qb_stats_df['qb'].tolist()
qb_list = []
for name in names:
    split_name = name.split(".")
    drop_second = split_name[0]
    clean_name = drop_second[:-1]
    qb_list.append(clean_name)
stats_df['Player'] = qb_list
#stats_df.head()
stats_df.sort_values('Player', ascending=True)
```

```python
stats_df = stats_df.sort_values('Player', ascending=True)
stats_df2 = stats_df[stats_df['Player'] ==""]
combined_names = []
uncleaned_names_list = stats_df2['qb'].tolist()
for uncleaned_name in uncleaned_names_list:
    split_name = uncleaned_name.split()
    split_name.remove(split_name[1])
    combined_name = " ".join(split_name)
    combined_names.append(combined_name)
stats_df.loc[stats_df2.index,'Player'] = combined_names
stats_df.sort_values('Player', ascending=True)
```

```python
#Create a game count column that will be used later for a weighted average.
game_count = stats_df['Player'].value_counts()
```

```python
ej = stats_df.loc[stats_df['Player'] == "E.J. Manuel"]
print(ej)
```

```python
#Drop the qb column and group the stats by Player
#Calculate averages of the stats and save to a new df
qb_stats_df = stats_df.drop('qb',axis=1)
new_df = qb_stats_df.groupby(['Player'])
avg_df = new_df["att","cmp","yds","ypa","td","int","lg","sack","loss","rate","game_points"].mean().round(2)
"E.J. Manuel" in avg_df.index
```

```python
#Combine the draft df and the average stats df 
#Add the game total to the combined df
combined_df = pd.merge(draft_df_qb,avg_df, on="Player", how="left")
combined_df = combined_df.dropna(how="any")
combined_df['Game Total'] = game_count
combined_df
```

```python
#Clean up and normalize the column names in the combined df
combined_df = combined_df.rename(index=str, columns={"Year": "Year Drafted", "Rnd": "Round Drafted", "Pick": "Overall Pick", "Pos": "Draft Position", 
                                              "att": "Avg Attempts", "cmp": "Avg Completions", "yds": "Avg Passing Yards", 
                                              "ypa": "Avg Yards per Attempt", "td": "Avg TDs", "sack": "Avg Sacks", 
                                              "loss": "Avg Loss of Yards", "rate": "Avg QBR", 
                                              "game_points": "Avg Points", "Game Total": "Game Total"})
combined_df
```

```python
#Create a connection to the in memory database (also creates the database)
#Convert the combined df to a table (called qb_stats) in the database
#Reflect the database to verfy creation
engine = create_engine('sqlite:///nfl_etl.sqlite', echo=False)
combined_df.to_sql('qb_stats', con=engine)
engine.execute("SELECT * FROM qb_stats").fetchall()
```

```python
#connect to the database
#Because the previously created table was created without a primary key...
#the following code duplicates the table into another table with a primary key and drops the initial table
import sqlite3
conn = sqlite3.connect('nfl_etl.sqlite')
c = conn.cursor()

c.executescript('''
    PRAGMA foreign_keys=off;

    BEGIN TRANSACTION;
    ALTER TABLE qb_stats RENAME TO old_table;
    
    /*create a new table with the same column names and types while
    defining a primary key for the desired column*/
    CREATE TABLE qb_stats2 (Player TEXT PRIMARY KEY NOT NULL,
                            Year_Drafted INTEGER, Round_Drafted INTEGER,
                            Overall_Pick INTEGER, Draft_Position TEXT, 
                            Avg_Attempts REAL, Avg_Completions REAL, 
                            Avg_Passing_Yards REAL,Avg_Yards_per_Attempt REAL, 
                            Avg_TDs REAL, Avg_Sacks REAL, Avg_Loss_of_Yards REAL,
                            Avg_QBR_REAL, Avg_Points REAL, Game_Total INTEGER);

    INSERT INTO qb_stats2 SELECT * FROM old_table;



    DROP TABLE old_table;
    COMMIT TRANSACTION;
    
    PRAGMA foreign_keys=off''')

#close out the connection
c.close()
conn.close()
```

```python
from sqlalchemy.ext.automap import automap_base
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
#query the database to reflect the new table to verify the data has been properly duplicated
engine.execute("SELECT * FROM qb_stats2").fetchall()
```
